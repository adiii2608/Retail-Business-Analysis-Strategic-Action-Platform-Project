```python
# %% [markdown]
# # Top Items Forecast (Robust Version)
# Uses Prophet if available, otherwise ARIMA (pmdarima) or statistical fallback.
# Inputs:
#   ./outputs_top_items/top10_products_monthly.csv
#   ./outputs_top_items/top10_action_plan_base.csv
# Outputs:
#   ./outputs_top_items_forecast/top10_forecast.csv
#   ./outputs_top_items_forecast/top10_forecast_action_plan.csv
#   (with method labels & safety stock suggestion)
#
# Columns expected in monthly file:
#   YearMonth (datetime), Description, Month_Quantity, Month_Value
#
# Safety Stock = std(last_3_months) * SAFETY_MULTIPLIER (default 1.0)
# Suggested_Stock = Forecast_Quantity + Safety_Stock
#
# ------------------------------------------------------------------

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings

plt.style.use("seaborn-v0_8-whitegrid")
sns.set_context("talk")
warnings.filterwarnings("ignore")

# Directories
INPUT_DIR = Path("./outputs_top_items")
OUTPUT_DIR = Path("./outputs_top_items_forecast")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

MONTHLY_FILE = INPUT_DIR / "top10_products_monthly.csv"
ACTION_BASE_FILE = INPUT_DIR / "top10_action_plan_base.csv"

SAFETY_MULTIPLIER = 1.0  # tune this (e.g., 1.28 for ~80% service level)
FORECAST_PERIODS = 1     # forecast next 1 month

# %%
# Try importing Prophet (optional)
try:
    from prophet import Prophet
    PROPHET_AVAILABLE = True
except Exception:
    PROPHET_AVAILABLE = False

# Try importing pmdarima (optional ARIMA)
try:
    from pmdarima import auto_arima
    ARIMA_AVAILABLE = True
except Exception:
    ARIMA_AVAILABLE = False

print(f"Prophet available: {PROPHET_AVAILABLE}")
print(f"ARIMA available: {ARIMA_AVAILABLE}")

# %%
# Load data
df_monthly = pd.read_csv(MONTHLY_FILE, parse_dates=['YearMonth'])
print("Monthly data shape:", df_monthly.shape)
display(df_monthly.head())

# Sanity: ensure correct columns
required_cols = {'YearMonth', 'Description', 'Month_Quantity'}
missing = required_cols - set(df_monthly.columns)
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# %%
# Helper Functions

def forecast_prophet(item_df, periods=1):
    """Forecast with Prophet; returns yhat for last period or raises Exception."""
    ts = item_df[['YearMonth', 'Month_Quantity']].rename(
        columns={'YearMonth': 'ds', 'Month_Quantity': 'y'}
    )
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        seasonality_mode="additive"
    )
    model.fit(ts)
    future = model.make_future_dataframe(periods=periods, freq='M')
    fc = model.predict(future)
    out = fc.tail(periods)
    return float(out['yhat'].iloc[-1]), "Prophet"

def forecast_arima(item_df, periods=1):
    """Forecast with auto_arima; returns forecast value."""
    ts = item_df.sort_values('YearMonth')['Month_Quantity'].astype(float).values
    model = auto_arima(ts, seasonal=True, m=12, suppress_warnings=True, error_action='ignore')
    fc = model.predict(n_periods=periods)
    return float(fc[-1]), "ARIMA"

def forecast_statistical(item_df):
    """
    Statistical fallback:
      - If >=3 months: mean of last 3
      - Else if >=1 month: last value
      - Else 0
    """
    qty = item_df.sort_values('YearMonth')['Month_Quantity'].tolist()
    if len(qty) >= 3:
        return float(np.mean(qty[-3:])), "3M_Average"
    elif len(qty) >= 1:
        return float(qty[-1]), "Last_Value"
    return 0.0, "Zero_Fallback"

def compute_safety_stock(item_df):
    """Std dev of last up to 3 months * multiplier (fallback 0)."""
    qty = item_df.sort_values('YearMonth')['Month_Quantity'].tail(3)
    if qty.shape[0] >= 2:
        return float(qty.std(ddof=0) * SAFETY_MULTIPLIER)
    else:
        return 0.0

# %%
# Forecast Loop
results = []

items = df_monthly['Description'].unique()
for item in items:
    sub = df_monthly[df_monthly['Description'] == item].copy()
    sub = sub.sort_values('YearMonth')

    # Skip too short series entirely (<2 points)
    if sub.shape[0] < 2:
        print(f"Skipping {item}: not enough data points.")
        continue

    forecast_value = None
    method_used = None

    # Try Prophet
    if PROPHET_AVAILABLE and sub.shape[0] >= 4:
        try:
            forecast_value, method_used = forecast_prophet(sub, periods=FORECAST_PERIODS)
        except Exception as e:
            print(f"[Prophet fail] {item}: {e}")

    # Try ARIMA if Prophet not used or failed
    if (forecast_value is None) and ARIMA_AVAILABLE and sub.shape[0] >= 4:
        try:
            forecast_value, method_used = forecast_arima(sub, periods=FORECAST_PERIODS)
        except Exception as e:
            print(f"[ARIMA fail] {item}: {e}")

    # Statistical fallback
    if forecast_value is None:
        forecast_value, method_used = forecast_statistical(sub)

    # Clean negative forecasts
    forecast_value = max(0, forecast_value)

    safety_stock = compute_safety_stock(sub)
    suggested_stock = forecast_value + safety_stock

    results.append({
        'Description': item,
        'History_Months': sub.shape[0],
        'Forecast_Quantity': round(forecast_value, 2),
        'Forecast_Method': method_used,
        'Safety_Stock': round(safety_stock, 2),
        'Suggested_Stock': round(suggested_stock, 2)
    })

forecast_df = pd.DataFrame(results).sort_values('Forecast_Quantity', ascending=False)
display(forecast_df.head())

# Save forecast results
forecast_df.to_csv(OUTPUT_DIR / "top10_forecast.csv", index=False)
print("Saved: top10_forecast.csv")

# %%
# Merge with base action plan (peaks & averages)
action_base = pd.read_csv(ACTION_BASE_FILE)

action_plan = action_base.merge(forecast_df, on='Description', how='left')

# Add reorder signal example: reorder if Suggested_Stock > Last_Month_Quantity
if 'Last_Month_Quantity' in action_plan.columns:
    action_plan['Reorder_Flag'] = np.where(
        (action_plan['Suggested_Stock'] > action_plan['Last_Month_Quantity'].fillna(0)),
        1, 0
    )
else:
    action_plan['Reorder_Flag'] = 1  # default

# Rank by Forecast or Value (if available)
if 'Avg_Monthly_Value' in action_plan.columns:
    action_plan['Priority_Rank'] = action_plan['Avg_Monthly_Value'].rank(method='dense', ascending=False)
else:
    action_plan['Priority_Rank'] = action_plan['Forecast_Quantity'].rank(method='dense', ascending=False)

action_plan = action_plan.sort_values('Priority_Rank')
action_plan.to_csv(OUTPUT_DIR / "top10_forecast_action_plan.csv", index=False)

display(action_plan.head())

print("Saved: top10_forecast_action_plan.csv")

# %%
# Visualization: Forecasted vs Last Month (if available)
if 'Last_Month_Quantity' in action_plan.columns:
    viz = action_plan[['Description', 'Last_Month_Quantity', 'Forecast_Quantity']].copy()
    viz_melt = viz.melt(id_vars='Description', value_name='Quantity', var_name='Type')

    plt.figure(figsize=(12,6))
    sns.barplot(data=viz_melt, x='Quantity', y='Description', hue='Type', palette='viridis')
    plt.title("Last Month vs Forecast Quantity (Top Items)")
    plt.xlabel("Quantity")
    plt.ylabel("Product")
    plt.tight_layout()
    plt.show()

# %%
# Forecast method distribution
plt.figure(figsize=(6,4))
sns.countplot(data=forecast_df, x='Forecast_Method', palette='Set2')
plt.title("Forecast Methods Used")
plt.ylabel("Count of Items")
plt.tight_layout()
plt.show()

# %%
print("Files written to:", OUTPUT_DIR.resolve())
for f in OUTPUT_DIR.iterdir():
    print(" -", f.name)

print("\nNEXT STEPS:")
print("1. Load 'top10_forecast_action_plan.csv' into Power BI for the Top Products Planning page.")
print("2. Optionally build GUI app to interactively inspect each item's history & forecast.")
print("3. Add inventory data (if available) to refine reorder logic.")
```

```python
# %% [markdown]
# # Top Items Analysis
# **Goal:** Identify Top 10 products (by sales value & quantity) over the last 12 months,
# build monthly time series, detect peak sales months, and create action planning files.
#
# **Outputs:**
# - top10_by_value.csv
# - top10_by_quantity.csv
# - top10_products_union.csv  (combined unique top items)
# - top10_products_monthly.csv
# - top10_peak_months.csv
# - top10_action_plan_base.csv
#
# These feed the forecasting + promotion planning notebook later.

# %%
# 1. Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import calendar

# Plot style
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_context("talk")

# %%
# 2. Configuration
DATA_PATH = Path(r"E:\c drive\amazon\data\online_retail\online_retail_II.xlsx")
OUTPUT_DIR = Path("./outputs_top_items")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# If you want to force a rolling 12-month window:
ROLLING_MONTHS = 12  # set to None to use full dataset

# %%
# 3. Load Data
df = pd.read_excel(DATA_PATH, parse_dates=['InvoiceDate'])

print("Raw shape:", df.shape)
df.head()

# %%
# 4. Basic Cleaning
# Remove rows with missing key fields
df = df.dropna(subset=['Description', 'Quantity', 'Price', 'InvoiceDate'])

# Keep only positive quantities & prices
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]

# Standardize description text (optional)
df['Description'] = df['Description'].str.strip()

# Add Sales Value
df['SalesValue'] = df['Quantity'] * df['Price']

print("Cleaned shape:", df.shape)
df[['Description', 'Quantity', 'Price', 'SalesValue']].head()

# %%
# 5. Filter to Last 12 Months (if dataset spans > 12 months)
max_date = df['InvoiceDate'].max()
min_date = df['InvoiceDate'].min()
print(f"Data range: {min_date.date()} -> {max_date.date()}")

if ROLLING_MONTHS:
    cutoff = max_date - pd.DateOffset(months=ROLLING_MONTHS)
    df_12m = df[df['InvoiceDate'] >= cutoff].copy()
    if df_12m['InvoiceDate'].nunique() < 30:  # fallback if too narrow
        print(" Not enough data in last 12 months filter. Using full dataset instead.")
        df_12m = df.copy()
else:
    df_12m = df.copy()

print("Filtered shape:", df_12m.shape)

# %%
# 6. Aggregate at Product Level (over the filtered period)
product_agg = (
    df_12m.groupby('Description')
          .agg(Total_Value=('SalesValue', 'sum'),
               Total_Quantity=('Quantity', 'sum'),
               Orders=('Invoice', 'nunique'),
               Avg_Price=('Price', 'mean'))
          .reset_index()
)

# Sort & extract Top 10
top10_value = product_agg.sort_values('Total_Value', ascending=False).head(10).copy()
top10_quantity = product_agg.sort_values('Total_Quantity', ascending=False).head(10).copy()

print("Top 10 by Value:")
display(top10_value)

print("Top 10 by Quantity:")
display(top10_quantity)

# %%
# 7. Save Top 10 Lists
top10_value.to_csv(OUTPUT_DIR / "top10_by_value.csv", index=False)
top10_quantity.to_csv(OUTPUT_DIR / "top10_by_quantity.csv", index=False)

# Union (unique set of product names from both lists)
top_items_union = pd.DataFrame(
    sorted(set(top10_value['Description']).union(set(top10_quantity['Description']))),
    columns=['Description']
)
top_items_union.to_csv(OUTPUT_DIR / "top10_products_union.csv", index=False)
print("Unique count of top products (union):", top_items_union.shape[0])

# %%
# 8. Build Monthly Time Series for Top Items Union
df_12m['YearMonth'] = df_12m['InvoiceDate'].dt.to_period('M').dt.to_timestamp()

top_monthly = (
    df_12m[df_12m['Description'].isin(top_items_union['Description'])]
    .groupby(['YearMonth', 'Description'])
    .agg(Month_Value=('SalesValue', 'sum'),
         Month_Quantity=('Quantity', 'sum'),
         Orders=('Invoice', 'nunique'))
    .reset_index()
)

top_monthly.to_csv(OUTPUT_DIR / "top10_products_monthly.csv", index=False)
print("Monthly time series rows:", top_monthly.shape[0])
top_monthly.head()

# %%
# 9. Peak Month Identification & Promotion Prep Month
peak_rows = []
for item in top_items_union['Description']:
    item_ts = top_monthly[top_monthly['Description'] == item]
    if item_ts.empty:
        continue
    # Determine peak month by quantity (could also use value)
    peak_row = item_ts.loc[item_ts['Month_Quantity'].idxmax()]
    peak_month_ts = peak_row['YearMonth']
    peak_month_num = peak_month_ts.month
    promo_month_num = peak_month_num - 1 if peak_month_num > 1 else 12
    
    peak_rows.append({
        'Description': item,
        'Peak_Month_Number': peak_month_num,
        'Peak_Month_Name': calendar.month_name[peak_month_num],
        'Promo_Preparation_Month_Number': promo_month_num,
        'Promo_Preparation_Month_Name': calendar.month_name[promo_month_num]
    })

peak_df = pd.DataFrame(peak_rows)
peak_df.to_csv(OUTPUT_DIR / "top10_peak_months.csv", index=False)
peak_df.head()

# %%
# 10. Merge for Base Action Plan
# Optionally include last observed month quantity & average monthly quantity
latest_month = top_monthly['YearMonth'].max()
recent_snapshot = (
    top_monthly[top_monthly['YearMonth'] == latest_month]
    .rename(columns={'Month_Quantity': 'Last_Month_Quantity',
                     'Month_Value': 'Last_Month_Value'})
    [['Description', 'Last_Month_Quantity', 'Last_Month_Value']]
)

avg_snapshot = (
    top_monthly.groupby('Description')
    .agg(Avg_Monthly_Quantity=('Month_Quantity', 'mean'),
         Avg_Monthly_Value=('Month_Value', 'mean'),
         Months_Available=('Month_Quantity', 'count'))
    .reset_index()
)

action_plan = (top_items_union
               .merge(peak_df, on='Description', how='left')
               .merge(recent_snapshot, on='Description', how='left')
               .merge(avg_snapshot, on='Description', how='left'))

action_plan.to_csv(OUTPUT_DIR / "top10_action_plan_base.csv", index=False)
display(action_plan)

# %%
# 11. Visualizations (Optional)

# Bar: Top 10 by Value
plt.figure(figsize=(10,6))
sns.barplot(data=top10_value, y='Description', x='Total_Value', palette='viridis')
plt.title("Top 10 Products by Sales Value")
plt.xlabel("Total Sales Value")
plt.ylabel("Product")
plt.tight_layout()
plt.show()

# Bar: Top 10 by Quantity
plt.figure(figsize=(10,6))
sns.barplot(data=top10_quantity, y='Description', x='Total_Quantity', palette='magma')
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Total Units Sold")
plt.ylabel("Product")
plt.tight_layout()
plt.show()

# Line plots (small multiples) for monthly quantity trends
n_items = top_items_union.shape[0]
cols = 2
rows = int(np.ceil(n_items / cols))
fig, axes = plt.subplots(rows, cols, figsize=(16, rows*3), sharex=True)
axes = axes.flatten()

for ax, item in zip(axes, top_items_union['Description']):
    sub = top_monthly[top_monthly['Description'] == item].sort_values('YearMonth')
    ax.plot(sub['YearMonth'], sub['Month_Quantity'], marker='o')
    ax.set_title(item[:40])  # truncate long names
    ax.set_ylabel("Qty")
for ax in axes[n_items:]:
    ax.axis('off')

fig.suptitle("Monthly Quantity Trend - Top Items", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

# %%
# 12. Summary Printouts
print("Files written to:", OUTPUT_DIR.resolve())
for f in OUTPUT_DIR.iterdir():
    print(" -", f.name)

print("\nNEXT STEP:")
print("Use 'top10_products_monthly.csv' and 'top10_action_plan_base.csv' in the forecasting notebook to predict next month's demand and finalize promotion scheduling.")
```
